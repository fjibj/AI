```python
#分别获得多个.docx文件中汉字的字频，合并取较高的字频并排序
# -*- coding: cp936 -*-
import docx
from functools import reduce

#读入.docx文件
filenames = ["8月售出活字统计.docx","10月销售字.docx"]

def GetWordDict(filename):
    file = docx.Document(filename)
    #获得.docx文件的文本内容
    fileIn = '\n\n'.join([
        paragraph.text for paragraph in file.paragraphs
    ])
    #print (fileIn)    

    word_lst = []
    word_dict = {}

    exclude_str = ".，。！？、（）【】<>《》=：+-*—“”…0123456789" 
    
    # 添加每一个字到列表中
    for line in fileIn:
        for char in line:
            word_lst.append(char)

    # 用字典统计每个字出现的个数       
    for char in word_lst:
        if char not in exclude_str:
            if char.strip() not in word_dict: # strip去除各种空白
                word_dict[char] = 1
            else :
                word_dict[char] += 1
    
    #print (word_dict)
    return word_dict

#合并dict,对key相同的取较大值
def GetMaxDict(dict1,dict2):
    for key in dict2:
        if dict1.get(key):
            dict1[key]=max(dict1[key],dict2[key])
        else:
            dict1[key]=dict2[key]
    return dict1

word_dicts = []
for filename in filenames:
    word_dicts.append(GetWordDict(filename))

w_dict = reduce(GetMaxDict,word_dicts)

with open("doc-out.txt",'w',encoding='UTF-8') as fileOut:
    # 排序
    #   x[1]是按字频排序，x[0]则是按字排序
    lstWords = sorted(w_dict.items(), key=lambda x:x[1],  reverse=True) 

    # 输出结果 (前100)
    print ('字符\t字频')
    print ('=============')
    for e in lstWords:
        print ('%s\t%d' % e)
        fileOut.write('%s, %d\n' % e)
      
     
```
