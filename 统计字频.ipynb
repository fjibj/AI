```python
#coding:utf-8

word_lst = []
word_dict = {}

exclude_str = ".，。！？、（）【】<>《》=：+-*—“”…0123456789" 

with open("10.txt","r") as fileIn ,open("10-out.txt",'w') as fileOut:
    
    # 添加每一个字到列表中
    for line in fileIn:
        for char in line:
            word_lst.append(char)

    # 用字典统计每个字出现的个数       
    for char in word_lst:
        if char not in exclude_str:
            if char.strip() not in word_dict: # strip去除各种空白
                word_dict[char] = 1
            else :
                word_dict[char] += 1

    # 排序
    #   x[1]是按字频排序，x[0]则是按字排序
    lstWords = sorted(word_dict.items(), key=lambda x:x[1],  reverse=True) 
   
    # 输出结果 (前100)
    print ('字符\t字频')
    print ('=============')
    for e in lstWords[:100]:  #输出前100个字
        print ('%s\t%d' % e)
        fileOut.write('%s, %d\n' % e)
```

```python
# 读取.docx文档，按倒序列出字频，并写到-doc-out.txt文件中
# -*- coding: cp936 -*-
import docx

#读入.docx文件
file = docx.Document("10月销售字.docx")  

#获得.docx文件的文本内容
fileIn = '\n\n'.join([
    paragraph.text for paragraph in file.paragraphs
])
#print (fileIn)    

word_lst = []
word_dict = {}

exclude_str = ".，。！？、（）【】<>《》=：+-*—“”…0123456789" 

with open("10-doc-out.txt",'w',encoding='UTF-8') as fileOut:
    
    # 添加每一个字到列表中
    for line in fileIn:
        for char in line:
            word_lst.append(char)

    # 用字典统计每个字出现的个数       
    for char in word_lst:
        if char not in exclude_str:
            if char.strip() not in word_dict: # strip去除各种空白
                word_dict[char] = 1
            else :
                word_dict[char] += 1

    # 排序
    #   x[1]是按字频排序，x[0]则是按字排序
    lstWords = sorted(word_dict.items(), key=lambda x:x[1],  reverse=True) 
   
    # 输出结果 (前100)
    print ('字符\t字频')
    print ('=============')
    for e in lstWords:
        print ('%s\t%d' % e)
        fileOut.write('%s, %d\n' % e)
         
```
